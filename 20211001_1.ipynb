```python
import random
import time

mywin=0
comwin=0

while True:
    num=random.randrange(1,4)
    if num==1:
      mywin=mywin+1
    elif num==3:
      comwin=comwin+1
    print("김효중:{0}.컴퓨터:{1}".format(mywin,comwin))
    time.sleep(1)
```
